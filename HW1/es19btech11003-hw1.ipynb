```python
#1a
import matplotlib.pyplot as plt
import numpy as np
def func(p):
    print("value of p:",p)
# 100 linearly spaced numbers
    x = np.linspace(-1,1,100)

# the function is here
    y = (1-abs(x)**p)**(1/p)
    y1=-(1-abs(x)**p)**(1/p)
# setting the axes at the centre
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.spines['left'].set_position('center')
    ax.spines['bottom'].set_position('center')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

# plot the function
    plt.plot(x,y,'b')
    plt.plot(x,y1, 'b')

# show the plot
    plt.show()
```

```python
#for p>=1
func(1)
func(1.5)
func(2)
func(10)
```

From the above plots we can clearly see that if we join any 2 points in the plot the line segment is definitely lying inside the figure so for all p>=1 the plot is convex

```python
#for 0<p<1
func(0.2)
func(0.5)
func(0.7)
func(0.9)
```

```python
#2a
import numpy as np                                                         
import matplotlib.pyplot as plt
import math
def func_2a(n):                                                                   
    A = []    
    B = []
    for i in range(0,100):
        a = i/100
        A.append(a)
  
    for i in A:
        if i>0.5 and i<1:
            B.append(1)
        elif i<=0.5 and i>=0:
            B.append((2*i+0.001)**n)

    plt.scatter(A,B)
    plt.show()
func_2a(1)
func_2a(math.inf)
```

```python
#3a
def func_3a(PMF):
    p=0
    for x in PMF:
        if x!=0:
            p-=x*(np.log2(x))
    return p
```

```python
def func_3b(p):
    if p==0 or p==1:
        return 0
    y = -(p)*(np.log2(p))-(1-p)*(np.log2(1-p))
    return y
count=0
a1=[]
a2=[]
while count<=1:
    l1=float(count)
    l2=func_3b(count)
    a1.append(l1)
    a2.append(l2)
    count+=0.001
m=max(a2)
print("value of p at which plot attained its maximum:",a1[a2.index(m)])
print("Maximum entropy value is:",m)
plt.plot(a1,a2)
plt.show()
```

```python
#6a
#function which takes joint PMF,index, RV as input and gives out appropriate conditional PMF
def func_6a(J_PMF,index,RV):
    a=len(J_PMF)
    b=len(J_PMF[0])
    A=[]
    sum=0
    if index==0:
        for i in range(b):
            sum+=J_PMF[RV][i]
        for i in range(b):
            J_PMF[RV][i]=J_PMF[RV][i]/sum
            A.append(float(J_PMF[RV][i]))
    elif index==1:
        for i in range(a):
            sum+=J_PMF[i][RV]
        for i in range(a):
            J_PMF[i][RV]=J_PMF[i][RV]/sum
            A.append(float(J_PMF[i][RV]))
    return A
m = int(input('number of rows, m = '))
n = int(input('number of columns, n = '))
matrix = []
# initialize the number of rows
for i in range(0,m):
    matrix += [0]
# initialize the matrix
for i in range (0,m):
    matrix[i] = [0]*n
for i in range (0,m):
    for j in range (0,n):
        #print ('entry in row: ',i+1,' column: ',j+1)
        matrix[i][j] = float(input())
index=int(input())
RV=int(input())
func_6a(matrix,index,RV)
```

```python
#6b
#Outputs the conditional entropy taking joint PMF,conditional RV
import numpy as np
def func_6b(J_PMF,index):
    J_PMF_copy=np.copy(J_PMF)
    a=len(J_PMF)
    b=len(J_PMF[0])
    A=[]
    sum=0;
    if index==0:
        for j in range(a):
            for i in range(b):
                sum+=J_PMF[j][i]
            for i in range(b):
                J_PMF[j][i]=J_PMF[j][i]/sum
            sum=0        
    elif index==1:
        for j in range(b):
            for i in range(a):
                sum+=J_PMF[i][j]
            for i in range(a):
                J_PMF[i][j]=J_PMF[i][j]/sum
            sum=0
    for i in range(a):
        for j in range(b):
            if J_PMF[i][j]!=0:
                J_PMF[i][j]=np.log2(J_PMF[i][j])
    multi_sum=0
    for i in range(a):
        for j in range(b):
            multi_sum+=float(J_PMF[i][j]*J_PMF_copy[i][j])
    return multi_sum
    
m = int(input('number of rows, m = '))
n = int(input('number of columns, n = '))
matrix = []
# initialize the number of rows
for i in range(0,m):
    matrix += [0]
# initialize the matrix
for i in range (0,m):
    matrix[i] = [0]*n
for i in range (0,m):
    for j in range (0,n):
        #print ('entry in row: ',i+1,' column: ',j+1)
        matrix[i][j] = float(input())
index=int(input())
func_6b(matrix,index)
```

```python
#outputs the value of D(p||q)
#7a
def func_7a(PMF1,PMF2):
    p=0
    for x in range(len(PMF1)):
        if PMF1[x]!=0 and PMF2[x]!=0:
            p+=PMF1[x]*(np.log2(PMF1[x]/PMF2[x]))
    return p
PMF1=[]
PMF2=[]
n=int(input("Number of values does both the RV take:"))
for i in range(0,n):
    l=float(input())
    PMF1.append(l)
print("values of 2nd RV")
for i in range(0,n):
    l=float(input())
    PMF2.append(l)
k=func_7a(PMF1,PMF2)
print("D(p||q):",k)
```

```python
#7d
def func_7d(pix_val,pix_val1):
    elements_count =[0]*256
    elements_count1 =[0]*256
# iterating over the elements for frequency
    for element in pix_val:
        elements_count[element] += 1
    sum1=sum(elements_count)
    elements_count = np.divide(elements_count,sum1)
    for element in pix_val1:
        elements_count1[element] += 1
    sum2=sum(elements_count1)
    elements_count1 = np.divide(elements_count1,sum2)
    return func_7a(elements_count,elements_count1)
          
im = Image.open('left.png', 'r')
pix_val = list(im.getdata())
im = Image.open('right.png', 'r')
pix_val1 = list(im.getdata())
print("D(p||q):",func_7d(pix_val,pix_val1))
print("D(q||p):",func_7d(pix_val1,pix_val))
```
